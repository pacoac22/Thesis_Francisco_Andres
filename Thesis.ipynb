# 4. Calculate Abnormal Reuturns 

Abnormal returns are the unexpected performance of an investment, in this case the companies stocks, compared to the expected risk free rate and market risk.

1 - Calculation of the returns of the datasets.

2 - Calculation of the abnormal returns.

## 4.1 Calculation of the returns of the datasets.

Use Log returns for the even study

```python
# Returns of the Emerging Markets
Emerging_Markets_returns = np.log(df_bfill_Emerging_Markets/ df_bfill_Emerging_Markets.shift(axis=1))
Emerging_Markets_returns = Emerging_Markets_returns.iloc[:,1:]

# Returns of the Developed Markets
Developed_Markets_returns = np.log(df_bfill_Developed_Markets/ df_bfill_Developed_Markets.shift(axis=1))
Developed_Markets_returns = Developed_Markets_returns.iloc[:,1:]

# Returns of the Treasury Bill and Index
TB_Index_returns = np.log(df_prices_TB_Index/ df_prices_TB_Index.shift(axis=1))
TB_Index_returns = TB_Index_returns.iloc[:,1:]

TB_Index_returns
```

```python
# Risk free returns and markets returns variables
TB_returns = TB_Index_returns.iloc[0,:]
Market_returns = TB_Index_returns.iloc[1:]

# Variables for the calculation of Abnormal Returns
Market_excess_return = Market_returns.subtract(TB_returns, axis=1)
Market_excess_return_array = Market_excess_return.values.reshape(-1, 1)
```

## 4.2 Calculation of the abnormal returns.


Calucalate the returns of the stock that exceed what is predicted around the event day using previous performance on the market 

```python
# Function to calculate the betas of the stocks in the market (CAPM

def alphas_and_betas(Stocks_excess_return, Market_excess_return_array):
    
    alphas = [] 
    betas = []
    reg = linear_model.LinearRegression()

    for Company, Stock_returns in Stocks_excess_return.iterrows():
        
        Stock_returns = Stock_returns.values.reshape(-1, 1)
        alphas.append(reg.fit(Stock_returns, Market_excess_return_array).intercept_)
        betas.append(reg.fit(Stock_returns, Market_excess_return_array).coef_)
    
    betas = np.array(betas)
    alphas = np.array(alphas)
    
    alphas_and_betas_df = pd.DataFrame({'Alpha': alphas.ravel(),'Beta': betas.ravel()}, index=Stocks_excess_return.index)
    
    return alphas_and_betas_df
```

```python
# Function to calculate the abnormal return of the different Markets

def abnormal_returns(Stocks_returns, TB_returns, Market_excess_return):
    
    Stocks_excess_return = Stocks_returns.subtract(TB_returns, axis=1)
    Alphas_Betas_Stocks_excess_returns = alphas_and_betas(Stocks_excess_return, Market_excess_return_array)
    
    betas = Alphas_Betas_Stocks_excess_returns['Beta']
    alphas = Alphas_Betas_Stocks_excess_returns['Alpha']
    
    expected_returns = np.outer(betas,Market_excess_return.values) #Multiplication of betas and excess market returns
    expected_returns = pd.DataFrame(expected_returns, index = Stocks_returns.index, columns = Stocks_returns.columns)
    expected_returns = expected_returns.add(alphas, axis='index') #Intercept + B*Excess Market Rertunrs
    
    abnormal_returns = (Stocks_excess_return - expected_returns)

    return abnormal_returns
```

```python
# Creating the DataFrames of the Abnormal Returns

AR_Emerging_Markets = abnormal_returns(Emerging_Markets_returns, TB_returns, Market_excess_return)
AR_Developed_Markets = abnormal_returns(Developed_Markets_returns, TB_returns, Market_excess_return)

AR_Emerging_Markets
```

# 5.  Calculation of AR & CAR

1 - Calculation of the Average Abnormal Returns over different days before and after the event

2 - Calculation of the Cumulative Abnormal Returns over different days before and after the event

## 5.1 Average Abnormal Returns

First, its calculated the Abnormal Returns for the companies over the event days and after that, the overage of the companies in the dates

```python
# Analysis of Average Abnormal Returns
def stats_AAR_companies(Abnormal_Retuns, Days_before, Days_after):
    
    df_event_study = Abnormal_Retuns.loc[:,Days_after:-Days_before]
    
    # Calculate the Mean and Standard Deviation of the AAR
    mean_AR = df_event_study.mean(axis = 0)
    std_AR = df_event_study.std(axis = 0)

    # Put everything in Dataframes
    stats = pd.DataFrame(mean_AR, columns=['Mean AR'])
    stats['STD AR'] = std_AR
    stats['T-Test'] = mean_AR / std_AR
    stats['P-Value'] = scyst.norm.cdf(stats['T-Test'])

    return stats
```

```python
# Abnormal Returns for +10 -10 days

display(stats_AAR_companies(AR_Developed_Markets, 10, 10))
display(stats_AAR_companies(AR_Emerging_Markets, 10, 10))
```

## 5.2 Cumulative Average Abnormal Returns

The Cumulative Abnormal Returns is calculated as the sum of the abnormal returns of the each company over the event dates

```python
# Analisis of Cumulative Average Abnormal Returns
def stats_CAAR_companies(Abnormal_Returns, Days_before, Days_after):
    
    # Extract the relevant event study window and calculate CAR, CAAR and its values
    AR_event_study = Abnormal_Returns.loc[:, Days_after:-Days_before]
    CAR = AR_event_study.sum(axis=1)
    
    mean_CAAR = np.mean(CAR)
    std_CAAR = np.std(CAR)
    t_test = mean_CAAR / std_CAAR
    p_value = scyst.norm.cdf(t_test)
    
    # Create DataFrame
    index = f' CAAR +{Days_before} / -{Days_after} days'
    stats = pd.DataFrame({
        'Mean CAAR': [mean_CAAR],
        'STD CAAR': [std_CAAR],
        'T-Test': [t_test],
        'P-Value': [p_value]
    },index = [index])

    return stats
```

```python
# Calculate CAAR for Developed Markets

CAAR_10_10 = stats_CAAR_companies(AR_Developed_Markets, 10, 10)
CAAR_5_5 = stats_CAAR_companies(AR_Developed_Markets, 5, 5)
CAAR_1_1 = stats_CAAR_companies(AR_Developed_Markets, 1, 1)

CAAR_Developed_Markets = pd.concat([ CAAR_10_10, CAAR_5_5, CAAR_1_1])
CAAR_Developed_Markets
```

# 6. Robust test

Using the GRANK-T Test with Standardized Abnormal Returns (SAR) & Standardized Cumulative Abnormal Returns (SCAR)

```python
def GRANK_T_test(Abnormal_Returns, Days_before, Days_after):
    
    # Calculation of SAR
    n = len(Abnormal_Returns)
    Abnormal_Returns = Abnormal_Returns.loc[:,10:-150]
    SAR = Abnormal_Returns.div(Abnormal_Returns.std(axis=1), axis=0)
    SAR_event_study = SAR.loc[:, Days_after:-Days_before]

    # Calculation of SCAR
    CAR = SAR_event_study.sum(axis=1)
    SCAR = CAR / np.std(CAR)
    
    # Calculation of GSAR
    SCAR_mean = np.mean(SCAR)
    SCAR_std = np.std(SCAR)
    Re_stand_SCAR = (SCAR - SCAR_mean) / SCAR_std
    
    #GSAAR & GRANK calculation
    GSAR = pd.concat([Re_stand_SCAR.to_frame(),SAR.loc[:, -Days_before-1:]],axis=1)
    GSAAR = GSAR.mean().to_frame(name='CAAR')
    Ranks = GSAAR.rank()
    L = len(GSAR)
    T = len(Ranks)
    U = (Ranks / (L + 1)) - 0.5 
    Z = U.loc[0] / np.sqrt(((U.mean()**2)*len(GSAR))/T)
    
    # Calculation of the GRANK-T statistic
    GRANK_T = Z / np.sqrt((T - 2) / (T - 1 - (Z ** 2)))
    p_value = 2 * (1 - scyst.norm.cdf(abs(GRANK_T)))
    
    return SCAR_mean, p_value
```

```python
grank_t_CAAR, p_value = GRANK_T_test(AR_Emerging_Markets, 1, 1)
print(f"GRANK-T CAAR: {grank_t_CAAR}, p-value: {p_value}")
grank_t_CAAR, p_value = GRANK_T_test(AR_Developed_Markets, 1, 1)
print(f"GRANK-T CAAR: {grank_t_CAAR}, p-value: {p_value}")
```

```python
# Add information for the regression data

# Add sector column to the Abrnormal Returns dataframe
common_index = AR_Emerging_Markets.index.intersection(df_Emerging_Markets.index)
AR_Emerging_Markets.loc[common_index, 'Sector Number'] = df_Emerging_Markets.loc[common_index, 'Sector Number']
common_index = AR_Developed_Markets.index.intersection(df_Developed_Countries.index)
AR_Developed_Markets.loc[common_index, 'Sector Number'] = df_Developed_Countries.loc[common_index, 'Sector Number']

df_Regression_data = pd.read_excel("Database.xlsx",'CAAR Regression',index_col = 0)
df_Regression_data
```

```python
# Caculate SCAR for window event

def SCAR(Abnormal_Returns, Days_before, Days_after):
    
    # Calculation of SAR
    n = len(Abnormal_Returns)
    Abnormal_Returns = Abnormal_Returns.loc[:,10:-150]
    SAR = Abnormal_Returns.div(Abnormal_Returns.std(axis=1), axis=0)
    SAR_event_study = SAR.loc[:, Days_after:-Days_before]

    # Calculation of SCAR
    CAR = SAR_event_study.sum(axis=1)
    SCAR = CAR / np.std(CAR)
    
    return SCAR
```

```python
# Adding SCAR and country to the regression dataset
SCAR_EM = pd.DataFrame(SCAR(AR_Emerging_Markets, 10, 10), columns=['SCAR'])
SCAR_EM['Country'] = AR_Emerging_Markets['Country']
SCAR_EM['Sector'] = AR_Emerging_Markets['Sector Number']
display(len(SCAR_EM))
SCAR_DM = pd.DataFrame(SCAR(AR_Developed_Markets, 10, 10), columns=['SCAR'])
SCAR_DM['Country'] = AR_Developed_Markets['Country']
SCAR_DM['Sector'] = AR_Developed_Markets['Sector Number']

Company_variables_dataset = pd.concat([SCAR_DM,SCAR_EM])

# Adding rest of country variables to the regression dataset
Regression_dataset = pd.merge(Company_variables_dataset, df_Regression_data, left_on='Country', right_index=True, how='left')

# Changing the country to be Emerging Markets variable
Regression_dataset.iloc[:561, Regression_dataset.columns.get_loc('Country')] = 1
Regression_dataset.iloc[561:, Regression_dataset.columns.get_loc('Country')] = 0
Regression_dataset
```
